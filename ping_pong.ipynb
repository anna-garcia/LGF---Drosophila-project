```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import cm
import matplotlib as mpl
from matplotlib import colormaps
from matplotlib.colors import LinearSegmentedColormap
import pandas as pd
import csv
from matplotlib.colors import LogNorm, Normalize
```

```python
df_multi = pd.read_csv("/home/anna/Documents/new_good_tables/pingpong_zscore_multimapping.tsv",decimal=',', sep='\t', header=0)
df_local = pd.read_csv("/home/anna/Documents/new_good_tables/pingpong_zscore_localmapping.tsv",decimal=',', sep='\t', header=0)
```

```python
fig = plt.figure(figsize=(10,5))

ax = plt.subplot2grid((3, 3), (0, 0), rowspan=3, colspan=1)
bx = plt.subplot2grid((3, 3), (0, 2), rowspan=3, colspan=2)

ax.set_prop_cycle('color',plt.cm.viridis(np.linspace(0,1,3)))
bx.set_prop_cycle('color',plt.cm.viridis(np.linspace(1,2,3)))

x = df_multi.Overlap_Size.to_list()


names = ['D. moj Y(rep1)','D. moj Y(rep2)','D. ari Y(rep 2)']

for i in df_multi[['z_y1_moj','z_y2_moj']]:
        column = df_multi[i].to_list()
        ax.plot(x,column, alpha=0.6, zorder=2)
        ax.legend(names, loc="upper right", prop={'size': 9})

ax.axhline(0, color='red', lw=0.9, zorder=0)
ax.set_ylabel('z-score', rotation=90)
ax.set_xlabel('Overlap Size (nt)')
ax.set_xlim(xmin=0, xmax=30)
ax.grid(alpha=0.8,linestyle='--', zorder=0)

for j in df_multi[['z_y2_ari']]:
        column = df_multi[j].to_list()
        bx.plot(x,column, alpha=1, zorder=2)
        bx.legend(['D. ari Y(rep 2)'],loc="upper right", prop={'size': 9})

bx.axhline(0, color='red', lw=0.9, zorder=0)
bx.set_ylabel('z-score', rotation=90)
bx.set_xlabel('Overlap Size (nt)')
bx.set_xlim(xmin=0, xmax=30)
bx.grid(alpha=0.8,linestyle='--', zorder=0)


plt.savefig('/home/anna/Documents/pingpong_multimapped.png', dpi=1200)

plt.show
```
