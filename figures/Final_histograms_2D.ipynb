```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib as mpl
from matplotlib import colormaps
import pandas as pd
import csv
from scipy.interpolate import make_interp_spline
```

```python
ovary = pd.read_csv("/home/anna/Downloads/Project/Length_Histograms_Ovaries.tsv",decimal=',', sep='\t', header=0)
testis = pd.read_csv("/home/anna/Downloads/Project/Length_Histograms_Testis.tsv", decimal=',', sep='\t', header=0)
chr_y = pd.read_csv("/home/anna/Downloads/Project/Length_Histograms_Ychr.tsv", decimal=',', sep='\t', header=0)
```

```python
fig = plt.figure(figsize=(8,5))
ax = plt.subplot()

ax.set_prop_cycle('color',plt.cm.magma(np.linspace(0,1,5)))

x = ovary.Reads_Length.to_list()
names = ['D. mojavensis 1','D. mojavensis 2','D. arizonae 1','D. arizonae 2']

for i in ovary:
    if i != 'Reads_Length':
        column = ovary[i].to_list()
        plt.plot(x,column)

ax.set_xlabel('Reads Length')
ax.set_xticks(x)

ax.set_ylabel('RPM', rotation=90)
ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useOffset=False)


plt.legend(names)
plt.grid(alpha=0.8)

plt.savefig('/home/anna/Downloads/Ovary_histogram_2D.png', dpi=1200)

plt.show
```
